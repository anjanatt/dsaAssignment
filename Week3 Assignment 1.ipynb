## Reading the given dataset:iris.xls

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter("ignore")
data=pd.read_excel('iris.xls')
data
```

## Displaying the columns

```python
data.columns
```

## Calculating mean values

```python
data.mean()
```

## Checking Null Values

```python
data.isnull()
```

## Visualization

```python
plt.figure(figsize=(12,12))
sns.boxplot(x=data['PL'], y=data['PW'])
plt.show
```

```python
plt.figure(figsize=(12,12))
sns.boxplot(x=data['SL'], y=data['SW'])
plt.show
```

```python
plt.figure(figsize=(12,12))
sns.barplot(x=data['PL'], y=data['PW'])
plt.show
```

```python
sns.set_style(style='white')
sns.pairplot(data, hue='Classification', size=2.5);
```

```python
plt.figure(figsize=(10,10))
sns.stripplot(data['Classification'],data['PW'])
```
