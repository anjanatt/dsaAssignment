## 2.  Total sales generated by all the zones for each month

```python
#Total sales generated by Zones in each month
data['Total Sales'] = data[['Zone - A', 'Zone - B', 'Zone - C', 'Zone - D']].sum(axis=1)
data[['Month','Total Sales']]
```

```python
from scipy.stats import f_oneway
stat, pvalue = f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
print('Stats = ',stat)
print('P-Value = ',pvalue)
```


## Checking that whether all the zones generate the same amount of sales.

```python
if pvalue < 0.05:
    print('Rejected Null Hypothesis ')
else:
    print('Alternate Hypothesis rejected')
```
